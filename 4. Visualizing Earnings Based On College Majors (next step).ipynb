## Scatter Plot

```python
recent_grads.plot("Sample_size", "Median", kind="scatter")
```

From the scatter plot above, we can see that there is no correlation between Sample_size and Median. But we can see that the lower sample_size has more variety of median and also had a chance of greater Median.

```python
recent_grads.plot("Sample_size", "Unemployment_rate", kind="scatter")
```

From the scatter plot above, we can see that there is no correlation between Sample_size and Unemployment_rate. But we can see that the lower sample_size has more variety of Unemployment_rate and also had a chance of greater Unemployment_rate and lesser Unemployment_rate.

```python
recent_grads.plot("Full_time", "Median", kind="scatter")
```

From the scatter plot above, we can see that there is no correlation between Full_time and Median. But we can see that the lower Full_time has more variety of Median and also had a chance of greater Median.

```python
recent_grads.plot("ShareWomen", "Unemployment_rate", kind="scatter")
```

From the scatter plot above, we can see that there is no correlation between ShareWomen and Unemployment_rate.

```python
recent_grads.plot("Men", "Median", kind="scatter")
```

From the scatter plot above, we can see that there is no correlation between Men and Median. But we can see that the lower Men has more variety of Median and also had a chance of greater Median.

```python
recent_grads.plot("Women", "Median", kind="scatter")
```

From the scatter plot above, we can see that there is no correlation between Woman and Median. But we can see that the lower Woman has more variety of median and also had a chance of greater Median.

## Histogram

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
recent_grads[cols].hist(bins=25, ax=ax, grid=False, xrot=40)
plt.show()
```

From the histogram above, we can see that there are more major that predominantly female and the most common median salary range is between 25000 to 50000.

## Bar Plots

```python
fig = plt.figure(figsize=(10,5))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)
recent_grads.head(10).plot(x="Major", y="ShareWomen", ax=ax1, kind="bar", legend=False)
recent_grads.tail(10).plot(x="Major", y="ShareWomen", ax=ax2, kind="bar", legend=False)
plt.show()
```
